```python
import pandas as pd
```

```python
data=pd.read_csv('CardioVascular.csv')
```

```python
data
```

```python
col=data.columns.to_list()
col
```

```python
sbp=[data['sbp(mm Hg)'].to_list()]
ldl=[data['ldl(mmol/L)'].to_list()]
adip=[data['adiposity'].to_list()]
tob=[data['tobacco(in Kg)'].to_list()]
alc=[data['alcohol'].to_list()]
typa=[data['typea'].to_list()]
age=[data['age'].to_list()]
fh=[data['FamHist'].to_list()]
hd=[data['CHD'].to_list()]
```

```python
dic={
col[0]:sbp,
col[1]:ldl,
col[2]:adip,
col[3]:tob,
col[4]:alc,
col[5]:typa,
col[6]:age,
col[7]:fh,
col[8]:hd
}
```

```python
dic
```

```python
data=pd.DataFrame(dic)
data
```
